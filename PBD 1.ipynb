```python
import numpy as np
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot
```

```python
df=pd.read_csv('jsrt_metadata.csv')
df=df.dropna()
df
```

```python
plt.pyplot.title("Total patients for diagnosis")
plt.pyplot.xlabel("Disease")
plt.pyplot.ylabel("Patients")
plt.pyplot.hist(df['diagnosis'].value_counts())
```

# 2

```python
each_disease=df.groupby("state")["diagnosis"].count()
```

```python
each_disease.plot(kind='pie')
```

# 3

```python
df_female= df.loc[df["gender"] == "Female"]
df_female.groupby("diagnosis")["gender"].count().plot(kind = "pie", figsize = (10,10),autopct='%0.1f%%')
```

```python
df_male= df.loc[df["gender"] == "Male"]
df_male.groupby("diagnosis")["gender"].count().plot(kind = "pie", figsize = (10,10),autopct='%0.1f%%')
```
